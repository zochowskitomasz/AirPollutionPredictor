```python
import requests
import pandas as pd
from datetime import datetime

API_KEY = "73d6d54ecae24663a6b72411252910"
BASE_URL = "https://api.worldweatheronline.com/premium/v1/past-weather.ashx"
CITY = "Warsaw"

def fetch_weather(start_date, end_date):
    params = {
        "key": API_KEY,
        "q": CITY,
        "date": start_date,
        "enddate": end_date,
        "tp": "1",
        "format": "json"
    }
    r = requests.get(BASE_URL, params=params)
    data = r.json()
    records = []
    for day in data["data"]["weather"]:
        date = day["date"]
        for hour in day["hourly"]:
            records.append({
                "datetime": f"{date} {hour['time'].zfill(4)[:2]}:00",
                "temperature_C": hour["tempC"],
                "humidity_%": hour["humidity"],
                "pressure_hPa": hour["pressure"],
                "wind_kph": hour["windspeedKmph"],
            })
    return pd.DataFrame(records)

# Zakres miesięczny 2024
ranges = [
    ("2025-01-01", "2025-01-31"),
    ("2025-02-01", "2025-02-29"),
    ("2025-03-01", "2025-03-31"),
    ("2025-04-01", "2025-04-30"),
    ("2025-05-01", "2025-05-31"),
    ("2025-06-01", "2025-06-30"),
    ("2025-07-01", "2025-07-31"),
    ("2025-08-01", "2025-08-31"),
    ("2025-09-01", "2025-09-30"),
    ("2025-10-01", "2025-10-31")
]

frames = [fetch_weather(start, end) for start, end in ranges]
df = pd.concat(frames)
df.to_csv("warsaw_weather_2025_hourly.csv", index=False)

print("Dane zapisano do warsaw_weather_2025_hourly.csv")
```
